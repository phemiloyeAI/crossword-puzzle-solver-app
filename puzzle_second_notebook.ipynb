```python
def calc_minimum_area(stats, n_labels):
    areas = []
    heights = []

    for i in range(1, n_labels):
        areas.append(stats[i, 4])
        heights.append(stats[i, 3])

    n = int(len(areas) * 0.1)

    sorted_area = sorted(areas)
    sorted_height = sorted(heights)

    del sorted_area[n]; del sorted_height[n]
    
    min_area = int(sum(sorted_area[:n+1])/n+1)
    min_height = int(sum(sorted_height[:n+1])/n+1)
    
    return min_area, min_height
```

```python
def load_image_array(puzzle_chars_dir, coordinates):
    img_array = []
    for coord in coordinates:
        coord = str(coord)
        c1, c2 = coord.split(",")

        coord = c1.strip()+","+c2.strip()

        path = puzzle_chars_dir+f"/{coord}.jpg"

        img = cv2.imread(path, 0)
        img_array.append(cv2.resize(img, (38,38), interpolation=cv2.INTER_AREA))

    img_array = np.stack(img_array)
    return img_array
```
